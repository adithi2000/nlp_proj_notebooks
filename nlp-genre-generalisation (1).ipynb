```python
COMMON_GENRES = [
    "drama",
    "comedy",
    "action",
    "romance",
    "thriller",
    "horror",
    "fantasy",
    "sci-fi",
    "crime",
    "mystery",
    "adventure",
    "history",
    "family"
]
```

```python
genre_map = {

    # ---------------- DRAMA ----------------
    "drama": "drama",
    "fiction": "drama",
    "novel": "drama",
    "world cinema": "drama",
    "non-fiction": "drama",

    # ---------------- COMEDY ----------------
    "comedy": "comedy",
    "romantic comedy": "comedy",

    # ---------------- ROMANCE ----------------
    "romance": "romance",
    "romance novel": "romance",
    "romance film": "romance",
    "romantic drama": "romance",

    # ---------------- ACTION ----------------
    "action": "action",
    "action/adventure": "action",

    # ---------------- ADVENTURE ----------------
    "adventure": "adventure",
    "adventure novel": "adventure",

    # ---------------- THRILLER ----------------
    "thriller": "thriller",
    "suspense": "thriller",

    # ---------------- HORROR ----------------
    "horror": "horror",

    # ---------------- SCI-FI ----------------
    "science fiction": "sci-fi",
    "speculative fiction": "sci-fi",
    "alternate history": "sci-fi",

    # ---------------- FANTASY ----------------
    "fantasy": "fantasy",

    # ---------------- CRIME ----------------
    "crime": "crime",
    "crime fiction": "crime",

    # ---------------- MYSTERY ----------------
    "mystery": "mystery",
    "detective fiction": "mystery",

    # ---------------- HISTORY ----------------
    "historical fiction": "history",
    "historical novel": "history",

    # ---------------- FAMILY ----------------
    "children's literature": "family",
    "family film": "family",
    "young adult literature": "family",

    # ---------------- SPECIAL CASES ----------------
    "animation": "family",
    "musical": "romance",
    "spy fiction": "thriller"
}
```

```python
#format removal
invalid_genres = {
    "black-and-white",
    "short film",
    "indie"
}
```

```python
def map_genres(genres):
    mapped = []
    for g in genres:
        g = g.lower().strip()
        
        if g in invalid_genres:
            continue
            
        if g in genre_map:
            mapped.append(genre_map[g])
    
    return list(set(mapped))
```

```python
df_book['mapped_genres'] = df_book['genres'].apply(map_genres)
df_movie['mapped_genres'] = df_movie['Genre'].apply(map_genres)
```
